```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from datetime import datetime

%matplotlib inline
```

```python
filename = "datasets/DC_Properties.csv"
df=pd.read_csv(filename)
df.head()
```

```python
df.corr()
```

```python
plt.figure(figsize=(15,10))
sns.violinplot(x='ROOMS', y='PRICE', data=df)
```

```python
nf = df[['HEAT','BATHRM','HF_BATHRM','BEDRM','ROOMS','LANDAREA','KITCHENS','FIREPLACES','PRICE','SALE_NUM','SALEDATE','STYLE','FULLADDRESS','YR_RMDL','WARD']]
nf.head()
```

```python
bins = [0,30000,60000,100000,250000,500000,100000000]
group_names = ['Affordable','Mid_range','Upper_mid','Affluent','Expensive','Ridiculous']
nf['Affordability'] = pd.cut(nf['PRICE'], bins, labels=group_names)
nf.head()
```

```python
nf_no_missing = nf.dropna()
nf_no_missing
```

```python
corr = nf.corr()

sns.heatmap(corr, vmin=-1, annot=True)
```

```python
nf.corr()
```

```python
plt.figure(figsize=(13,7))
sns.violinplot(x='Affordability',
               y='BEDRM',
               data=nf,
               inner=None, #removes bars inside violin plot
               )

plt.title('Land area by Affordability')
```

```python
plt.figure(figsize=(13,7))
sns.violinplot(x='WARD',
               y='PRICE',
               data=nf,
               inner=None, #removes bars inside violin plot
               )
```

```python
plt.figure(figsize=(20,15))
sns.violinplot(x='YR_RMDL',
               y='PRICE',
               data=df,
               inner=None, #removes bars inside violin plot
               )
```
